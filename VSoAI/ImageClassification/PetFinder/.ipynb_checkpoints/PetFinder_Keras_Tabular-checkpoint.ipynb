```python
# fill missing values

# fill missing categorical values
def fill_missing_cat(dataset,columns):
    for col in columns:
        dataset[col].fillna(value="missing", inplace=True)
    return (dataset)

# fill missing continuous values
def fill_missing_cont(dataset,columns):
    for col in columns:
        #dataset[col].fillna(value=0.0, inplace=True)
        dataset[col].fillna(value=dataset[col].mean(), inplace=True)
    return (dataset)
```

```python
def convertType(df, cont_vars, cat_vars):
    for v in cont_vars:
        df[v] = df[v].astype('float32') 

    for v in cat_vars:
        df[v] = df[v].astype('category').cat.as_ordered()
        #df[v] = df[v].astype('object')
```

```python
# collect max values for categorical and text columns

max_dict = {} # max values for each column

for col in cat_vars:
    max_dict[col] = np.max([train[col].max(), valid[col].max()])+1

max_dict
```

```python
# Fast.ai way of creating embedding size
# first calcualting cardinality size
cat_sz = [(c, len(df[c].cat.categories)+1) for c in cat_vars]
cat_sz
```

```python
# But here is Jeremy Howard's rule of thumb
emb_szs = [(c, min(50, (c+1)//2)) for _,c in cat_sz]
emb_szs
```

```python
# define keras variables
#from keras.preprocessing.sequence import pad_sequences

def get_keras_vars(dataset):
    X = {}
    for col in cat_vars:
        X[col] = np.array(dataset[col])
   
 #     for col in textcols:
 #         if verboseout:
 #             print("text col is",col)
 #         X[col] = pad_sequences(dataset[col], maxlen=max_dict[col])
        
    return X

X_train = get_keras_vars(train)
X_valid = get_keras_vars(valid)
X_test = get_keras_vars(test_df)
print("Through definition of keras variables")

type(X_train), type(X_valid), type(X_test)
```

```python
# Define & run model
# Identify model input, layers, and compilation details.

# basic imports
from tensorflow.keras.layers import Input, Dropout, Dense, BatchNormalization, Activation, concatenate, GRU, Embedding, Flatten, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import ModelCheckpoint, Callback, EarlyStopping
from tensorflow.keras import regularizers
#from tensorflow.keras.layers.normalization import BatchNormalization
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.optimizers import SGD
from tensorflow.keras import backend as K
#from tensorflow.keras.utils.vis_utils import plot_model
```

```python
embeddings = {}
catinputs = {}
inputlayerlist = []
collistfix = [] # list of categorical layer definitions
catemb = 10 # vector size for categorical column embeddings

# iterate through categorical columns and build input layer definitions and add to list of layers
for col in cat_vars:
        catinputs[col] = Input(shape=[1],name=col)
        inputlayerlist.append(catinputs[col])
        embeddings[col] = (Embedding(max_dict[col],catemb) (catinputs[col]))
        # batchnorm all 
        embeddings[col] = (BatchNormalization() (embeddings[col]))
        collistfix.append(embeddings[col])
```

```python
# seed main_l with few cat varaiables
# hyperparameters
dropout_rate = 0.003

main_l = concatenate([Dropout(dropout_rate) (Flatten() (embeddings['Name']) ),Dropout(dropout_rate) (Flatten() (embeddings['PetID']) )])
for cols in cat_vars:
    if (cols != 'Name') & (cols != 'PetID'):
        main_l = concatenate([main_l,Dropout(dropout_rate) (Flatten() (embeddings[cols]) )])
```

```python
main_l
```

```python
output = Dense(1, activation="hard_sigmoid") (main_l)

# define input and output for model: inputlayerlist is list built in iteration through features; output is defined via main_l layer list built up from layers defined for each features
model = Model(inputlayerlist, output)

# define optimizer: currently using Adam; also experimented with SGD
optimizer = Adam(lr=learning_rate)

model.compile(loss="binary_crossentropy", optimizer=optimizer, metrics=["accuracy"])
#model.summary()
```
