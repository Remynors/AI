# Create Image Data Generator

```python
w = h = 112
batch_size = 32
n_classes = len(clss)
```

```python
# # Flow from test train directories -> mine

train_datagen = keras.preprocessing.image.ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)

test_datagen = keras.preprocessing.image.ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        './train',
        target_size=(w, h),
        batch_size=batch_size,
        class_mode='categorical')

validation_generator = test_datagen.flow_from_directory(
        './test',
        target_size=(w, h),
        batch_size=batch_size,
        class_mode='categorical')
```

```python
n_train, n_test = 16464, 4116
```

# Create a Model

```python
base_model = keras.applications.Xception(
    include_top=False, 
    weights='imagenet', 
    input_tensor=keras.layers.Input(shape=(w, h, 3)),
)

x = base_model.output
x = keras.layers.GlobalAveragePooling2D()(x)
x = keras.layers.Dropout(0.5)(x)
preds = keras.layers.Dense(n_classes, activation='softmax')(x)
model = keras.models.Model(input=base_model.input, output=preds)
```

```python
for layer in base_model.layers:
    layer.trainable = False
```

```python
opt = keras.optimizers.Adam(lr=1e-4)
```

```python
model.compile(
    loss=keras.losses.categorical_crossentropy, 
    optimizer=opt, 
    metrics=[
        keras.metrics.categorical_crossentropy,
        keras.metrics.top_k_categorical_accuracy
    ],
)
```

```python
model_freezed = model.fit_generator(
    train_generator,
    steps_per_epoch = n_train // batch_size,
    epochs=10,
    validation_data=validation_generator,
    validation_steps=n_test // batch_size,
)
```

```python
model.save('gdrive/My Drive/dogs/freezed.h5')
```

```python
plt.figure(0)
plt.plot(model_freezed.history['top_k_categorical_accuracy'],'r')
plt.plot(model_freezed.history['val_top_k_categorical_accuracy'],'g')
plt.xticks(np.arange(0, 20, 1.0))
plt.rcParams['figure.figsize'] = (8, 6)
plt.xlabel("Num of Epochs")
plt.ylabel("Accuracy")
plt.title("Training Accuracy vs Validation Accuracy")
plt.legend(['train','validation'])
 
plt.figure(1)
plt.plot(model_freezed.history['loss'],'r')
plt.plot(model_freezed.history['val_loss'],'g')
plt.xticks(np.arange(0, 20, 1.0))
plt.rcParams['figure.figsize'] = (8, 6)
plt.xlabel("Num of Epochs")
plt.ylabel("Loss")
plt.title("Training Loss vs Validation Loss")
plt.legend(['train','validation'])

# plt.figure(2)
# plt.plot(model_history.history['mean_squared_error'],'r')
# plt.plot(model_history.history['val_mean_squared_error'],'g')
# plt.xticks(np.arange(0, 20, 1.0))
# plt.rcParams['figure.figsize'] = (8, 6)
# plt.xlabel("Num of Epochs")
# plt.ylabel("MSE")
# plt.title("Training Loss vs Validation Loss")
# plt.legend(['train','validation'])
 
plt.show()
```
